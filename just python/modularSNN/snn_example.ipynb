# A Python Module for Easily Building Self-Organized Spiking Neural Networks
#### This notebook will show how to use the snn python module to build multi-layered SNNs. For documentation, see `snn.py` or call `help()` on any class/method/function.


First, we will make a `Network` object to hold the layers of our network.

```python
model = snn.Network()
```

Next, we specify layers to add to our network with a name, number of neurons, and neurontype. We can call the LIFNeuron class without specifying parameters, which will use the default values or we can specify the parameters.

```python
layer1 = snn.Layer('Ret', 1600, snn.LIFNeuron(1600, ai=30, ao=10, ri=3, ro=6))
layer2 = snn.Layer('LGN', 512, snn.LIFNeuron(512, ai=30, ao=10, ri=1.2, ro=1.5, sqR=10, tau_v=1, tau_th=10, th_plus=6,v_th=1))
layer3 = snn.Layer('V1', 400, snn.LIFNeuron(400, thresh_mu =30, thresh_sig= 2, tau_v=1, tau_th=10, th_plus=6,v_th=1))
```

Now, we add these layers to our network, specifying the learning rule for the higher layers.

```python
model.add(layer1)
model.add(layer2, snn.STDP(.1))
model.add(layer3, snn.STDP(.15))
```

Now we run our model. We will run it for 30 seconds with a time step (dt) of .1 seconds.

```python
model.run(30)
```

To plot results, we can either plot all the plots of the network by calling `model.plot()` or we can plot a specific layer with `layer1.plot()`.

```python
layer1.plot()
```

Since in this case, there are a lot of neurons, we can specify how many neurons we want to plot.

```python
layer1.plot(10)
```

```python
model.plot(10)
```

Now, we can use the `plot_wave` method on individual layers to visualize the wave in that layer. 

```python
layer1.plot_wave()
```
